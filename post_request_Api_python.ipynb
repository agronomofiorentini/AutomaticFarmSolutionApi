# Automatic Farm Solution

Welcome to this tutorial to help you understand how to use the Automatic Farm Solution API.
In this tutorial we will introduce you the basic to produce prescription maps by using the automatic procedure of the Automatic Farm Solution API.

# GEOJson request

The following API call allow you to get a geojson that can be used as prescription map.

```python
# Import libraries

import requests
import fiona
import geopandas as gpd
from requests.auth import HTTPBasicAuth

# Define the url of the API

url = "https://www.api.automaticfarmsolutionwebapp.com/AFS/AutoPrescriptionMapGEOjson?AutoPrescription=False&NumberZone=3&userfertilizer=200&Strategy=highwherehigh"  # Replace with the actual URL for uploading

# Path to the GeoJSON file
file_path = "E:/new_plumber/county.geojson"

# Read the GeoJSON file contents
with open(file_path, "r") as file:
    geojson_data = file.read()

# Set the headers for the request
headers = {
    "Content-Type": "application/json"
}

# Make the POST request with the GeoJSON data as the request body
response = requests.post(url, 
                         data=geojson_data, 
                         headers=headers,
                         auth=HTTPBasicAuth('username', 'password'))

b = bytes(response.content)
with fiona.BytesCollection(b) as f:
    crs = f.crs
    gdf = gpd.GeoDataFrame.from_features(f, crs=crs)
    print(gdf.head())

    
#  Visualize the data   
gdf.explore("Zone")
```

# Convert geoJSON to Shapefile

```python
gdf.to_file('prescription_map.shp')
```
